```python
def up(l):
    if(0 in l[0:3]):
        return l
    else:
        ar=list(l)
        i = ar.index(0)
        ar[i-3],ar[i]=ar[i],ar[i-3]
        return tuple(ar)
```

```python
def down(l):
    if(0 in l[6:9]):
        return l
    else:
        ar = list(l)
        i = ar.index(0)
        ar[i+3],ar[i]=ar[i],ar[i+3]
        return tuple(ar)
```

```python
def left(l):
    if(l[0]==0 or l[3]==0 or l[6]==0):
        return l
    else:
        ar = list(l)
        i = ar.index(0)
        ar[i-1],ar[i]=ar[i],ar[i-1]
        return tuple(ar)
```

```python
def right(l):
    if(l[2]==0 or l[5]==0 or l[8]==0):
        return l
    else:
        ar = list(l)
        i = ar.index(0)
        ar[i+1],ar[i]=ar[i],ar[i+1]
        return tuple(ar)
```

```python
def nexts(x):
    temp = []
    if(x!=up(x)):
        temp.append(up(x))
    if(x!=down(x)):
        temp.append(down(x))
    if(x!=left(x)):
        temp.append(left(x))
    if(x!=right(x)):
        temp.append(right(x))
    return temp
```

```python
from collections import defaultdict
import heapq

class node:
    def __init__(self, state, parent=None, hc=True):
        self.state = state
        self.parent = parent
        self.h = self.manhattan(state)
        #self.h = self.out_of_sequence(state)
        if self.parent == None:
            self.g = 0
        else:
            self.g = parent.g + 1
        if hc == False:
            self.f = self.h
        else:
            self.f = self.g + self.h
            
    def __lt__(self, z):
        if self.f == z.f:
            return self.g < z.g
        return self.f < z.f

    def manhattan(self, st):
        m = 0
        for i in range(9):
            if st[i] == 0:
                continue
            m += abs(st[i] // 3 - i // 3) + abs(st[i] % 3 - i % 3)
        return m
    
    def out_of_sequence(self,st):
        m = 0
        for i in range(9):
            if(i==0 and st[i]!=0):
                m+=1
            elif(i!=8):
                if(st[i]+1 == st[i+1]):
                    continue
            else:
                if(st[i]==8):
                    continue
                m+=2
        return m

class solve:
    def __init__(self):
        self.states = defaultdict(list)

    def addEdge(self, u, v):
        self.states[u].append(v)
        
    def uninformed(self, start, goal):
        explored = set()
        queue = [[start]]
        while queue:
            route = queue.pop(0)
            node = route[-1]

            if node not in explored:
                neighbours = self.states[node]
                for neighbour in neighbours:
                    new_path = list(route)
                    new_path.append(neighbour)
                    queue.append(new_path)
                    if neighbour == goal:
                        return new_path, len(explored)
            explored.add(node)
        return "IMPOSSIBLE", 0
    
    def Greedy_BFS(self, start, goal):

        frontier = []
        explored = set()
        visited = set()

        heapq.heappush(frontier, node(start, hc=False))
        visited.add(start)

        while frontier:
            n = heapq.heappop(frontier)

            if tuple(n.state) in explored:
                continue

            explored.add(n.state)

            if n.state == goal:
                res = n
                l = len(visited)
                return res, l

            for i in nexts(n.state):
                if i not in visited:
                    z = node(i, n, hc=False)
                    heapq.heappush(frontier, z)
                    visited.add(i)
                elif i not in explored:
                    z = node(i, n, hc=False)
                    heapq.heappush(frontier, z)

        return None, None
    
    def A_star(self, start, goal):

        frontier = []
        explored = set()
        visited = set()
        f = 0
        heapq.heappush(frontier, node(start))
        visited.add(start)

        while frontier:
            n = heapq.heappop(frontier)

            if n.state in explored:
                continue

            explored.add(n.state)

            if n.state == goal:
                res = n
                l = len(visited)
                return res, l

            for i in nexts(n.state):
                if i not in visited:
                    z = node(i, n)
                    heapq.heappush(frontier, z)
                    visited.add(i)
                elif i not in explored:
                    z = node(i, n)
                    heapq.heappush(frontier, z)

        return None, None
    
    def find_path(self, res):
        x = list()

        while True:
            x.append(res.state)
            if (res.parent == None):
                break
            res = res.parent
        x.reverse()
        sn = 1
        for i in x:
            print()
            print(str(sn)+")")
            p(i)
            print()
            sn+=1
        return sn-1
```

```python
start = (7,2,4,
          5,0,6,
          8,3,1)
goal = (0,1,2,3,4,5,6,7,8)
```

```python
puzzle = solve()
```

```python
count = 0
q = set()
q.add(start)

while len(q) != 0:
    if(count%50000==0):
        print(count)
    x = q.pop()
    count+=1
    for i in nexts(x):
        puzzle.addEdge(x,i)
        if i not in puzzle.states.keys():
            if i not in q:
                q.add(i)
```

```python
print("\nUninformed Breadth First search")
path, nodes = puzzle.uninformed(start, goal)
if nodes >0:
    print("\n No of nodes Discovered : ",nodes)
    print()
    sn = 1
    for i in path:
        print()
        print(str(sn)+")")
        p(i)
        print()
        sn+=1
    print("No of moves required : ",len(path))
else:
    print(path)
```

```python
print("\nInformed Greedy Best First Search")
path, nodes = puzzle.Greedy_BFS(start, goal)
print("\nNo of nodes Discovered : ",nodes)
print()
moves = puzzle.find_path(path)
print("No of moves required : ",moves)
```

```python
print("\nInformed A* Search")
path, nodes = puzzle.A_star(start, goal)
print("\nNo of nodes Discovered : ",nodes)
moves = puzzle.find_path(path)
print("No of moves required : ",moves)
```
